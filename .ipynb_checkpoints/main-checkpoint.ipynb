```python
import scipy.io
import numpy as np
import matplotlib.pyplot as plt

# Load the training and testing data for digits 5 and 6
Numpyfile_training_5 = scipy.io.loadmat('training_data_5.mat')
Numpyfile_testing_5 = scipy.io.loadmat('testing_data_5.mat')
Numpyfile_training_6 = scipy.io.loadmat('training_data_6.mat')
Numpyfile_testing_6 = scipy.io.loadmat('testing_data_6.mat')

# Combine the data from both training and testing sets for digits 5 and 6
images_train_5 = Numpyfile_training_5['train_data_5']
images_train_6 = Numpyfile_training_6['train_data_6']
images_test_5 = Numpyfile_testing_5['test_data_5']
images_test_6 = Numpyfile_testing_6['test_data_6']

training_data_combined = np.concatenate((images_train_5, images_train_6), axis=0)
testing_data_combined = np.concatenate((images_test_5, images_test_6), axis=0)

# Calculate mean and standard deviation for training data
mean_train = np.mean(training_data_combined, axis=(1, 2))
std_deviation_train = np.std(training_data_combined, axis=(1, 2))

# Z score or standardization for training data
standardized_images_train = (training_data_combined - mean_train[:, np.newaxis, np.newaxis]) / std_deviation_train[:, np.newaxis, np.newaxis]

# Perform PCA for training data
mean_centered_data_train = standardized_images_train
num_samples_train, num_features_train = mean_centered_data_train.shape[0], mean_centered_data_train.shape[1]
cov_matrix_train = np.zeros((num_features_train, num_features_train))

for i in range(num_features_train):
    for j in range(i, num_features_train):
        cov_ij = np.sum(mean_centered_data_train[:, i] * mean_centered_data_train[:, j]) / (num_samples_train - 1)
        cov_matrix_train[i, j] = cov_ij
        cov_matrix_train[j, i] = cov_ij

eigenvalues_train, eigenvectors_train = np.linalg.eig(cov_matrix_train)

# Sort eigenvalues and eigenvectors for training data
sorted_indices_train = np.argsort(eigenvalues_train)[::-1]
eigenvalues_train = eigenvalues_train[sorted_indices_train]
eigenvectors_train = eigenvectors_train[:, sorted_indices_train]

k_train = 2  # Number of principal components to retain for training data
num_components_train = k_train

# Projection for training data
projection_train = np.dot(standardized_images_train, eigenvectors_train[:, :num_components_train])

# Class labels for training data
labels_train = [5] * images_train_5.shape[0] + [6] * images_train_6.shape[0]

# Calculate mean and standard deviation for testing data
mean_test = np.mean(testing_data_combined, axis=(1, 2))
std_deviation_test = np.std(testing_data_combined, axis=(1, 2))

# Z score or standardization for testing data
standardized_images_test = (testing_data_combined - mean_test[:, np.newaxis, np.newaxis]) / std_deviation_test[:, np.newaxis, np.newaxis]

# Perform PCA for testing data
mean_centered_data_test = standardized_images_test
num_samples_test, num_features_test = mean_centered_data_test.shape[0], mean_centered_data_test.shape[1]
cov_matrix_test = np.zeros((num_features_test, num_features_test))

for i in range(num_features_test):
    for j in range(i, num_features_test):
        cov_ij = np.sum(mean_centered_data_test[:, i] * mean_centered_data_test[:, j]) / (num_samples_test - 1)
        cov_matrix_test[i, j] = cov_ij
        cov_matrix_test[j, i] = cov_ij

eigenvalues_test, eigenvectors_test = np.linalg.eig(cov_matrix_test)

# Sort eigenvalues and eigenvectors for testing data
sorted_indices_test = np.argsort(eigenvalues_test)[::-1]
eigenvalues_test = eigenvalues_test[sorted_indices_test]
eigenvectors_test = eigenvectors_test[:, sorted_indices_test]

k_test = 3  # Number of principal components to retain for testing data
num_components_test = k_test

# Projection for testing data
projection_test = np.dot(standardized_images_test, eigenvectors_test[:, :num_components_test])

# Class labels for testing data
labels_test = [5] * images_test_5.shape[0] + [6] * images_test_6.shape[0]

# Create scatter plots for 2-D visualization of both training and testing data
plt.figure(figsize=(15, 6))

# Training Data Plot
plt.subplot(1, 2, 1)
unique_labels_train = set(labels_train)
colors_train = ['r', 'g', 'b', 'c', 'm', 'y', 'k']
for label in unique_labels_train:
    indices_train = [i for i, l in enumerate(labels_train) if l == label]
    x_train = projection_train[indices_train, 0]
    y_train = projection_train[indices_train, 1]
    plt.scatter(x_train, y_train, c=colors_train[label - 5], label=f'Training: Class {label}')
plt.title('2-D Projection of Training Samples (Digits 5 and 6)')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.legend()

# Testing Data Plot
plt.subplot(1, 2, 2)
unique_labels_test = set(labels_test)
colors_test = ['r', 'g', 'b', 'c', 'm', 'y', 'k']
for label in unique_labels_test:
    indices_test = [i for i, l in enumerate(labels_test) if l == label]
    x_test = projection_test[indices_test, 0]
    y_test = projection_test[indices_test, 1]
    plt.scatter(x_test, y_test, c=colors_test[label - 5], label=f'Testing: Class {label}')
plt.title('2-D Projection of Testing Samples (Digits 5 and 6)')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.legend()

plt.tight_layout()
plt.show()


training_data_class_5 = images_train_5.reshape(images_train_5.shape[0], -1)  # Flatten the data
training_data_class_6 = images_train_6.reshape(images_train_6.shape[0], -1)

# Calculate the mean and covariance matrix for class 5
mean_class_5 = np.mean(training_data_class_5, axis=0)
covariance_matrix_class_5 = np.cov(training_data_class_5, rowvar=False)

# Calculate the mean and covariance matrix for class 6
mean_class_6 = np.mean(training_data_class_6, axis=0)
covariance_matrix_class_6 = np.cov(training_data_class_6, rowvar=False)

print("Parameters for Class 5:")
print("Mean:\n", mean_class_5)
print("Covariance Matrix:\n", covariance_matrix_class_5)

print("Parameters for Class 6:")
print("Mean:\n", mean_class_6)
print("Covariance Matrix:\n", covariance_matrix_class_6)
```

```python
from scipy.stats import multivariate_normal

# Function to regularize a covariance matrix and ensure it is positive definite
def regularize_covariance(cov_matrix, reg_param):
    identity_matrix = np.identity(cov_matrix.shape[0])
    reg_cov_matrix = cov_matrix + reg_param * identity_matrix
    return reg_cov_matrix

# Regularize the covariance matrices
reg_param = 0.0005  # Adjust as needed
cov_matrix_class_5_reg = regularize_covariance(covariance_matrix_class_5, reg_param)
cov_matrix_class_6_reg = regularize_covariance(covariance_matrix_class_6, reg_param)

# Create multivariate normal distributions for each class using the regularized covariance matrices
pdf_class_5 = multivariate_normal(mean=mean_class_5, cov=cov_matrix_class_5_reg)
pdf_class_6 = multivariate_normal(mean=mean_class_6, cov=cov_matrix_class_6_reg)

# Function to perform classification and calculate accuracy
def classify_and_evaluate(data, true_labels):
    predicted_labels = []
    correct_count = 0

    for i, sample in enumerate(data):
        # Calculate the posterior probabilities for each class
        posterior_prob_class_5 = pdf_class_5.pdf(sample)
        posterior_prob_class_6 = pdf_class_6.pdf(sample)

        # Assign the sample to the class with the higher posterior probability
        predicted_class = 5 if posterior_prob_class_5 > posterior_prob_class_6 else 6
        predicted_labels.append(predicted_class)

        # Check if the prediction is correct
        if predicted_class == true_labels[i]:
            correct_count += 1

    # Calculate accuracy
    accuracy = correct_count / len(data)
    return accuracy, predicted_labels

# Classify the training set and calculate accuracy
accuracy_train, predicted_labels_train = classify_and_evaluate(training_data_combined.reshape(training_data_combined.shape[0], -1), labels_train)

# Classify the testing set and calculate accuracy
accuracy_test, predicted_labels_test = classify_and_evaluate(testing_data_combined.reshape(testing_data_combined.shape[0], -1), labels_test)

# Report the accuracy for the training and testing sets
print("Accuracy for the Training Set: {:.2f}%".format(accuracy_train * 100))
print("Accuracy for the Testing Set: {:.2f}%".format(accuracy_test * 100))
```
