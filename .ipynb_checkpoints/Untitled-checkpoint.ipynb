```python
import numpy as np
from scipy.io import loadmat
import scipy
from scipy.stats import multivariate_normal
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the training and testing data for digits 5 and 6
Numpyfile_training_5 = scipy.io.loadmat('training_data_5.mat')
Numpyfile_testing_5 = scipy.io.loadmat('testing_data_5.mat')
Numpyfile_training_6 = scipy.io.loadmat('training_data_6.mat')
Numpyfile_testing_6 = scipy.io.loadmat('testing_data_6.mat')

# Combine the data from both training and testing sets for digits 5 and 6
# Storing Loaded data into dictionaries
images_train_5 = Numpyfile_training_5['train_data_5']
images_train_6 = Numpyfile_training_6['train_data_6']
images_test_5 = Numpyfile_testing_5['test_data_5']
images_test_6 = Numpyfile_testing_6['test_data_6']

print(images_train_5.shape)
```

```python
# # Reshape the training data for digits 5 and 6
# images_train_5 = images_train_5.reshape(images_train_5.shape[0], -1)
# images_train_6 = images_train_6.reshape(images_train_6.shape[0], -1)

# print(images_train_5.shape, images_train_6.shape)

# # Reshape the testing data for digits 5 and 6
# images_test_5 = images_test_5.reshape(images_test_5.shape[0], -1)
# images_test_6 = images_test_6.reshape(images_test_6.shape[0], -1)

# print(images_test_5.shape, images_test_6.shape)
```

```python
#Now we vectorize the 28*28 array to 784 d . Basically flattening the 2d array to single array

X_zeros_train = np.reshape(images_train_5, (-1, images_train_5.shape[1] * images_train_5.shape[2]))
X_ones_train = np.reshape(images_train_6, (-1, images_train_6.shape[1] * images_train_6.shape[2]))
print(X_zeros_train.shape, X_ones_train.shape)
```

```python
X_zeros_test = np.reshape(images_test_5, (-1, images_test_5.shape[1] * images_test_5.shape[2]))
X_ones_test = np.reshape(images_test_6, (-1, images_test_6.shape[1] * images_test_6.shape[2]))
print(X_zeros_test.shape, X_ones_test.shape)
```

```python
X_train = np.concatenate((images_train_5, images_train_6), axis=0)
X_test = np.concatenate((images_test_5, images_test_6), axis=0)
print(X_train.shape, X_test.shape)
```

```python
y_train = np.concatenate((np.zeros(images_train_5.shape[0], dtype='int'), np.ones(images_train_6.shape[0], dtype='int')), axis=0)
y_test = np.concatenate((np.zeros(images_test_5.shape[0], dtype='int'), np.ones(images_test_6.shape[0], dtype='int')), axis=0)
print(y_train.shape, y_test.shape)
```

```python
def normalize(dataset):
    mean_vector = np.mean(dataset, axis=0)
    print("Shape of mean vector: {}".format(mean_vector.shape))
    std_vector = np.std(dataset, axis=0)
    print("Shape of std vector: {}".format(std_vector.shape))


    epsilon = 1e-8
    std_vector = np.where(std_vector == 0, epsilon, std_vector)

    
    scaled_dataset = (dataset - mean_vector)/std_vector
    print("Shape of normalized data: {}".format(scaled_dataset.shape))
    
    return scaled_dataset
```

```python
X_train_normalized = normalize(X_train)
```

```python
X_test_normalized = normalize(X_test)
```

```python
covariance_matrix = np.cov(X_train_normalized.T)
print(covariance_matrix.shape)
```

```python
eigen_values, eigen_vectors = np.linalg.eig(covariance_matrix)
print(eigen_values.shape, eigen_vectors.shape)
```

```python
eigen_pairs = [(np.abs(eigen_values[i]), eigen_vectors[:, i]) for i in range(eigen_values.shape[0])]
eigen_pairs.sort(key=lambda x: x[0], reverse=True)
```

```python
explained_variance_threshold = 0.95

total_variance = sum(eigen_values)
explained_variances = [(i/total_variance) for i in sorted(eigen_values, reverse=True)]
cumulative_variances = np.cumsum(explained_variances)

num_principal_components = 0

for i, exp_var in enumerate(cumulative_variances):
    if exp_var > explained_variance_threshold:
        num_principal_components = i + 1
        break

print("# of Key principal components explaining {}% of data: {}".format(explained_variance_threshold*100, num_principal_components))
```

```python
cumulative_variances[:2]

```

```python
cumulative_variances = np.real(cumulative_variances)

fig, axes = plt.subplots(1,1,figsize=(16,7), dpi=100)
plt.plot(cumulative_variances, color='firebrick')
plt.title('Variance Explained %', fontsize=22)
plt.xlabel('# of PCs', fontsize=16)
```

```python
projection_matrix = np.hstack((eigen_pairs[0][1].reshape(-1, 1), eigen_pairs[1][1].reshape(-1, 1)))
projection_matrix.shape
```

```python
projected_data = X_train_normalized.dot(projection_matrix)
projected_data.shape
```

```python
projected_test_data = X_test_normalized.dot(projection_matrix)
projected_test_data.shape
```

```python
train_df = pd.DataFrame(projected_data)
train_df['Y'] = y_train

train_df.head()
```

```python
test_df = pd.DataFrame(projected_test_data)
test_df['Y'] = y_test

test_df.head()
```

```python
mu_zero = np.mean(train_df[[0, 1]][train_df['Y'] == 0], axis=0)
print("Mean Vector for Class 5: ")
print(mu_zero)
```

```python
cov_mat_zero = np.cov(train_df[[0, 1]][train_df['Y'] == 0].T)
print("Covariance Matrix for Class 6: ")
print(cov_mat_zero)
```

```python
zero_distribution = multivariate_normal(mu_zero, cov_mat_zero)
```

```python
mu_one = np.mean(train_df[[0, 1]][train_df['Y'] == 1], axis = 0)
mu_zero = np.real(mu_zero)

print("Mean Vector for Class 1: ")
print(mu_one)
```

```python
cov_mat_one = np.cov(train_df[[0, 1]][train_df['Y'] == 1].T)
cov_mat_zero = np.real(cov_mat_zero)

print("Covariance Matrix for Class 1: ")
print(cov_mat_one)
```

```python
one_distribution = multivariate_normal(mu_one, cov_mat_one)
```

```python
p_x_zero = lambda x: zero_distribution.pdf(x)
p_x_one = lambda x: one_distribution.pdf(x)
```

```python

def classify(datapoint, p_zero=0.5, p_one=0.5):
        
    p_zero_x = p_x_zero(datapoint) * p_zero
    p_one_x = p_x_one(datapoint) * p_one
    
    if p_zero_x >= p_one_x:
        return 0
    return 1
```

```python
p_zero = X_zeros_train.shape[0]/X_train.shape[0]
p_one = X_ones_train.shape[0]/X_train.shape[0]
print(p_zero, p_one)
```

```python
p_zero = p_one = 0.5
```

```python
y_pred_train = [classify(i, p_zero=p_zero, p_one=p_one) for i in projected_data]
y_pred_test = [classify(i, p_zero=p_zero, p_one=p_one) for i in projected_test_data]
```

```python
train_df['Y_pred'] = y_pred_train
test_df['Y_pred'] = y_pred_test
```

```python
training_acc = np.mean(train_df['Y'] == train_df['Y_pred'])
print(training_acc)
```

```python
testing_acc = np.mean(test_df['Y'] == test_df['Y_pred'])
print(testing_acc)
```
